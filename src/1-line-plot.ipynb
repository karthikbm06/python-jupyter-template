### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Step 1: Create x-values and y-values
x_values = list(range(0, 11))  # x values from 0 to 10
y_values = [x**2 for x in x_values]  # y values as the square of x

# Step 2: Plot the line graph
plt.plot(x_values, y_values)

# Step 3: Customize the plot
plt.title('Basic Line Plot')  # Add title
plt.xlabel('x')  # Label x-axis
plt.ylabel('y')  # Label y-axis
plt.grid(True)  # Show grid

# Display the plot
plt.show()
```
