### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second,  chart in the fourth.
import matplotlib.pyplot as plt

# Data for the plots
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot a line chart in the first subplot
axs[0, 0].plot(x, y, color='blue', marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Plot a pie chart in the third subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 0].set_title('Pie Chart')

# Plot a scatter plot in the fourth subplot
axs[1, 1].scatter(x, y, color='purple')
axs[1, 1].set_title('Scatter Plot')
axs[1, 1].set_xlabel('X values')
axs[1, 1].set_ylabel('Y values')

# Adjust the layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```
