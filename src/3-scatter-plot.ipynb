### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Step 1: Define the data points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Step 2: Create the scatter plot
plt.scatter(x, y)

# Step 3: Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', ha='left', va='bottom')

    # Step 4: Add a title to the plot
    plt.title('Scatter Plot Example')

    # Display the plot
    plt.show()
```
